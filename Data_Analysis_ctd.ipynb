```python
#Luke Kurta
import pandas as pd

#resturants
df_rests = pd.read_csv(
    "../Project/team_project_dataset/restaurants_santa_barbara.csv",
    sep=",",
    encoding="latin-1",
    header=None
)

df_rests.columns = df_rests.iloc[0]
df_rests = df_rests[1:]
df_rests.reset_index(drop=True)

#test
df_test = pd.read_csv(
    "../Project/team_project_dataset/test_reviews_santa_barbara.csv",
    sep=",",
    encoding="latin-1",
    header=None
)

df_test.columns = df_test.iloc[0]
df_test = df_test[1:]
df_test.reset_index(drop=True)

#train
df_train = pd.read_csv(
    "../Project/team_project_dataset/train_reviews_santa_barbara.csv",
    sep=",",
    encoding="latin-1",
    header=None
)

df_train.columns = df_train.iloc[0]
df_train = df_train[1:]
df_train.reset_index(drop=True)
```

```python
num_restaurants = len(df_rests['business_id'].unique()) #number of rests
num_restaurants_in_data = len(pd.concat([df_train, df_test])['business_id'].unique()) #num of rests in review data
restaurants_in_data_list = pd.concat([df_train, df_test])['business_id'].unique() #list of rests in review data
#df_train_unique = df_train.drop_duplicates(subset="business_id")
#df_rests.merge(df_train_unique, on="business_id") --> gives df of the restaurants present in the train data

missing_reviews = (df_train["text"].isna().sum()) + (df_test["text"].isna().sum()) #check missing vals (none)
missing_ratings = (df_train["stars"].isna().sum()) + (df_test["stars"].isna().sum()) #check missing vals (none)
num_ratings_reviews = df_train.shape[0] + df_test.shape[0] #for all of this, it's nice to have the count for both train and test, just to analyze the dataset. of course, when we begin training, the only data that matters is the train data -- the test data is technically "unknown" during training and testing (we are trying to approximate the test data)
num_ratings_reviews_train = df_train.shape[0]

print(num_restaurants)
print(num_ratings_reviews_train)
```

```python
#restaurants by category -- lists of types of restaurants
asian_categories = ["asian","japanese","sushi","thai","chinese","indian","korean","ramen"]
asian_list = (df_rests.explode("categories").loc[lambda x: x["categories"].str.lower().isin(asian_categories), "business_id"].unique())
num_asian = len(asian_list)

niche_categories = ["argentine", "moroccan", "german", "latin american", "caribbean"]
niche_list = (df_rests.explode("categories").loc[lambda x: x["categories"].str.lower().isin(niche_categories), "business_id"].unique())
num_niche = len(niche_list)

mexican_categories = ["mexican", "tacos"]
mexican_list = (df_rests.explode("categories").loc[lambda x: x["categories"].str.lower().isin(mexican_categories), "business_id"].unique())
num_mexican = len(mexican_list)

italian_categories = ["italian", "pizza"]
italian_list = (df_rests.explode("categories").loc[lambda x: x["categories"].str.lower().isin(italian_categories), "business_id"].unique())
num_italian = len(italian_list)

american_categories = ["american", "cajun/creole", "barbeque", "burgers", "delis", "bagels", "soul food"]
american_list = (df_rests.explode("categories").loc[lambda x: x["categories"].str.lower().isin(american_categories), "business_id"].unique())
num_american = len(american_list)

nightlife_categories = ["beer", "nightlife", "bars", "pubs", "breweries", "sports"]
nightlife_list = (df_rests.explode("categories").loc[lambda x: x["categories"].str.lower().isin(nightlife_categories), "business_id"].unique())
num_nightlife = len(nightlife_list)

fastfood_categories = ["fast food", "deli"]
fastfood_list = (df_rests.explode("categories").loc[lambda x: x["categories"].str.lower().isin(fastfood_categories), "business_id"].unique())
num_fastfood = len(fastfood_list)

cafe_categories = ["cafes", "breakfast"]
cafe_list = (df_rests.explode("categories").loc[lambda x: x["categories"].str.lower().isin(cafe_categories), "business_id"].unique())
num_cafe = len(cafe_list)
```

```python
#let's do a similar thing, but now using the attributes. this requires a good bit of preprocessing, since the attributes column is in dictionary form but is all in a string --> we need to turn it from a string into a dictionary type, and then we also need to make sure we fix the values (the quotations are all over the place)

import ast
def str_to_dict(s):
    try:
        return ast.literal_eval(s)
    except: #need this or else it won't run
        return {}
df_rests['attributes_dict'] = df_rests['attributes'].apply(str_to_dict)

def clean_values(d):
    return {k: (v.strip("'\"") if isinstance(v, str) else v) for k, v in d.items()}
df_rests['attributes_dict'] = df_rests['attributes_dict'].apply(clean_values)
```

```python
#now that we have the attributes column all nice, we can filter

good_for_kids_df = df_rests[df_rests['attributes_dict'].apply( lambda d: d.get('GoodForKids') == 'True' and d.get('RestaurantsAttire') == 'casual' and d.get('Alcohol') == "u'none")]

easy_food_df = df_rests[df_rests['attributes_dict'].apply( lambda d: d.get('RestaurantsAttire') == 'casual' and d.get('RestaurantsTakeOut') == 'True' or d.get('RestaurantReservations') == 'False' and d.get('RestaurantsPriceRange2') == '1')]
```

```python
#dataframe containing review info about each restaurant -- one using train and test data, one using only train
merge1 = pd.concat([df_train, df_test]).merge(df_rests, how = 'left', on = 'business_id')

#review count for each
review_counts = merge1.groupby('business_id').count()
df_rests = df_rests.set_index('business_id')
df_rests['review_count'] = review_counts['review_id']
df_rests = df_rests.reset_index()
df_rests['review_count'] = df_rests['review_count'].fillna(0).astype(int)
```

```python
#average rating for each rest

merge1['stars'] = pd.to_numeric(merge1['stars'])
rating_avgs = merge1.groupby('business_id')['stars'].agg('mean').reset_index()
df_rests['avg_rating'] = df_rests['business_id'].map(rating_avgs.set_index('business_id')['stars'])
df_rests = df_rests.reset_index()
```

```python
#percentiles for ratings for each rest
rating_25th = merge1.groupby('business_id')['stars'].quantile(.25).reset_index()
rating_50th = merge1.groupby('business_id')['stars'].quantile(.50).reset_index()
rating_75th = merge1.groupby('business_id')['stars'].quantile(.75).reset_index()
df_rests['25th_percentile'] = df_rests['business_id'].map(rating_25th.set_index('business_id')['stars'])
df_rests['50th_percentile'] = df_rests['business_id'].map(rating_50th.set_index('business_id')['stars'])
df_rests['75th_percentile'] = df_rests['business_id'].map(rating_75th.set_index('business_id')['stars'])
df_rests = df_rests.reset_index()
```
